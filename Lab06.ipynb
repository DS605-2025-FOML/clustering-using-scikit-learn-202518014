# Clustering Using Scikit Learn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Part 1

```python
#loading the dataset
df = pd.read_csv('Mall_Customers.csv')
```

```python
#print the datatype, not null count and coulumn anems of each column to understand the number of null values 
df.info()
```

```python
#selecting only Annual Income and Spending Score
X = df[['Annual Income (k$)','Spending Score (1-100)']].copy()
```

```python
X.head()
```

```python
#plot the scatter plot to get a rought idea about the clusters and patterns in dataset
plt.figure(figsize=(8,6))
sns.scatterplot(x='Annual Income (k$)',y='Spending Score (1-100)', data=X, s=60)
plt.title('Annual Income v/s Spending Score')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

```python
#scaling the values
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### Part 2

```python
#finding optimal number of clusters
from sklearn.cluster import KMeans

wcss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

#ploting the wcss for k number of clusters
plt.figure(figsize=(8,6))
plt.plot(range(1,11), wcss, marker = 'o')
plt.title('Elbow Method')
plt.xlabel('Number of CLusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#selecting 5 as the optimal number of clusters
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
y_kmeans = kmeans.fit_predict(X)
centroids = kmeans.cluster_centers_

print('Custer Labels for First 10 Customers : ')
print(y_kmeans[:10])
```

```python
#Visualisation of KMeans
plt.figure(figsize=(8,6))
plt.scatter(X['Annual Income (k$)'], X['Spending Score (1-100)'],
            c=y_kmeans, cmap='rainbow', s=60, alpha=0.7, edgecolors='k')
#plotting the centroids
centers = kmeans.cluster_centers_
plt.scatter(centers[:,0], centers[:,1], c='black', s=200, marker='X',
            label='Centroids')
plt.title('Customer Segments (Using KMeans)')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```

### Part 3

```python
#creating a dendogram
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering

linked = linkage(X, method='ward')
plt.figure(figsize=(10,6))
dendrogram(linked)
plt.title('Dendogram for Mall Customers')
plt.xlabel('Customers')
plt.ylabel('Euclidiean Distances')
plt.show()
```

```python
#appling Agglomerative clustering with number of clusters as 5
optimal_clusters_hc = 5
hc_model = AgglomerativeClustering(n_clusters=optimal_clusters_hc,
                                   linkage='ward')
hc_labels = hc_model.fit_predict(X[['Annual Income (k$)','Spending Score (1-100)']])
X['HC_Cluster'] = hc_labels
```

```python
#Visualisation of Agglomerative Clustering
plt.figure(figsize=(10,6))
colors = ['red', 'blue', 'green','cyan', 'yellow']
for i in range(optimal_clusters_hc):
    plt.scatter(
        X[X['HC_Cluster']==i]['Annual Income (k$)'],
        X[X['HC_Cluster']==i]['Spending Score (1-100)'],
        s=60,
        c=colors[i],
        label = f'Cluster {i+1}'
    )
plt.title('Customer Segments (Using Agglomerative Clustering)')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```

### Part 4

```python
#Scale the features and create X_scaled DataFrame
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled_array = scaler.fit_transform(X[['Annual Income (k$)', 'Spending Score (1-100)']])
X_scaled = pd.DataFrame(X_scaled_array, columns=['Annual Income (k$)','Spending Score (1-100)'])
```

```python
from sklearn.cluster import DBSCAN

dbscan_model = DBSCAN(eps=5, min_samples=5)
dbscan_labels = dbscan_model.fit_predict(X_scaled[['Annual Income (k$)','Spending Score (1-100)']])

X_scaled['DBSCAN_Cluster'] = dbscan_labels
```

```python
#visualisation of DBSCAN
plt.figure(figsize=(8,6))
unique_labels = set(dbscan_labels)
colors = ['red', 'blue', 'green','cyan', 'magenta', 'yellow', 'purple']

for label in unique_labels:
    if label == -1:
        color = 'black'
        label_name = 'Noise'
    else:
        color = colors[label%len(colors)]
        label_name = f'Cluster {label+1}'

    plt.scatter(
        X_scaled[dbscan_labels == label]['Annual Income (k$)'],
        X_scaled[dbscan_labels == label]['Spending Score (1-100)'],
        s=60,
        c=color,
        label = label_name
    )
plt.title('Customer Segments (Using DBSCAN)')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```

```python
dbscan_model = DBSCAN(eps=0.45, min_samples=9)
dbscan_labels = dbscan_model.fit_predict(X_scaled[['Annual Income (k$)','Spending Score (1-100)']])

X_scaled['DBSCAN_Cluster'] = dbscan_labels
```

```python
#visualisation of DBSCAN
plt.figure(figsize=(8,6))
unique_labels = set(dbscan_labels)
colors = ['red', 'blue', 'green','cyan', 'magenta', 'yellow', 'purple']

for label in unique_labels:
    if label == -1:
        color = 'black'
        label_name = 'Noise'
    else:
        color = colors[label%len(colors)]
        label_name = f'Cluster {label+1}'

    plt.scatter(
        X_scaled[dbscan_labels == label]['Annual Income (k$)'],
        X_scaled[dbscan_labels == label]['Spending Score (1-100)'],
        s=60,
        c=color,
        label = label_name
    )
plt.title('Customer Segments (Using DBSCAN)')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```
